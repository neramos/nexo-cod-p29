```python
# discrete_vs_continuous_evolution.py
# Demonstrates how finite recursive steps approximate continuous physics

import numpy as np
import matplotlib.pyplot as plt

def continuous_evolution(t_max=10, dt=0.001):
    """Standard continuous differential equation (e.g., harmonic oscillator)"""
    t = np.arange(0, t_max, dt)
    x = np.sin(t)  # Analytic solution
    return t, x

def discrete_evolution(t_max=10, steps=100):
    """Finite recursive update (computational finitism)"""
    x = 0.0
    v = 1.0  # Initial velocity
    trajectory = []
    times = []

    dt = t_max / steps  # Finite time step (no limits)

    for i in range(steps):
        trajectory.append(x)
        times.append(i * dt)
        # Euler method (finite difference)
        x_new = x + v * dt
        v_new = v - x * dt  # Simple harmonic force
        x, v = x_new, v_new

    return np.array(times), np.array(trajectory)

# Comparison
t_cont, x_cont = continuous_evolution()
t_disc, x_disc = discrete_evolution(steps=50)  # Low resolution = visible discreteness

plt.plot(t_cont, x_cont, 'k-', label='Continuous Model (Idealization)')
plt.plot(t_disc, x_disc, 'ro-', label='Discrete Recursion (Physical Reality?)')
plt.title("Continuous vs. Discrete Evolution: Is Smoothness Real?")
plt.legend()
plt.show()
```
