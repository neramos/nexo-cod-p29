```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Generate Fibonacci numbers
def generate_fibonacci(max_n=10000):
    fib = [0, 1]
    while True:
        next_fib = fib[-1] + fib[-2]
        if next_fib > max_n:
            break
        fib.append(next_fib)
    return np.array(fib[1:])  # skip 0, start from 1

N_max = 1000000000000  # 10^12 - change if needed
fib_numbers = generate_fibonacci(N_max)

print(f"Generated {len(fib_numbers)} Fibonacci numbers")

# Polar spiral using Fibonacci numbers
values = fib_numbers.astype(float)
theta = values * np.log(values + 1)   # standard n * log(n)
r = values

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(13, 13), facecolor='black')
plt.scatter(x, y, c='#00ffcc', s=3, alpha=0.9)

plt.title(f"Polar Spiral using Fibonacci Numbers\n({len(fib_numbers)} terms up to ~{fib_numbers[-1]:,})",
          color='white', fontsize=15)
plt.axis('off')
plt.tight_layout()
plt.show()
```
