```python
"""
finitist_many_worlds_refutation.py
===================================
Direct refutation of infinite Many-Worlds.

Shows MWI must be FINITE, not infinite.

FIXED: Uses logarithms to avoid computational overflow
"""

import numpy as np
import matplotlib.pyplot as plt

# Constants
MAX_BITS = 1e122  # Maximum information in universe
PLANCK_TIME = 5.39e-44  # s

print("=" * 70)
print("REFUTING INFINITE MANY-WORLDS")
print("=" * 70)

def finite_many_worlds(max_bits=MAX_BITS, n_measurements=100):
    """
    Simulate Many-Worlds with FINITE information budget.

    FIXED: Use logarithms instead of calculating 2^(10^122) directly
    """
    bits_per_measurement = 1
    max_measurements = max_bits // bits_per_measurement

    # DON'T calculate 2^(10^122) directly - use log10 instead
    log10_branches_after_100 = n_measurements * np.log10(2)  # ≈ 30.1
    log10_branches_max = max_bits * np.log10(2)  # ≈ 3e121

    print(f"\nMany-Worlds Branching:")
    print(f"  Bits per measurement: {bits_per_measurement}")
    print(f"  Maximum measurements: {max_measurements:.2e}")
    print(f"  Branches after 100 measurements: 2^{n_measurements} ≈ 10^{log10_branches_after_100:.2f}")
    print(f"  Maximum branches: 2^{int(max_measurements):,} ≈ 10^{log10_branches_max:.2e}")
    print(f"  Time to exhaust all branches: {max_measurements * PLANCK_TIME / (365.25 * 24 * 3600):.2e} years")
    print(f"\n✓ Many-Worlds is FINITE, not infinite")
    print(f"✓ Branching STOPS when information budget exhausted")
    print(f"✓ This REFUTES infinite Many-Worlds interpretation")
    print(f"✓ Only FINITE computation tree exists")

    return max_measurements, log10_branches_after_100, log10_branches_max


def plot_many_worlds_refutation(max_measurements, log10_branches_100, log10_branches_max):
    """
    Plot Many-Worlds branching with finite limit.

    FIXED: Plot log10(branches) instead of actual branches
    """
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Left: Branching growth (log scale)
    measurements = np.arange(0, 200, 10)
    log10_branches = measurements * np.log10(2)  # log10(2^n) = n * log10(2)

    axes[0].plot(measurements, log10_branches, 'b-', linewidth=2, label='log₁₀(Branches)')
    axes[0].axhline(y=log10_branches_max, color='red', linestyle='--', linewidth=2,
                    label=f'Max log₁₀(branches) ≈ {log10_branches_max:.2e}')
    axes[0].set_title('Many-Worlds Branching Growth\n(Log Scale)', fontsize=12, fontweight='bold')
    axes[0].set_xlabel('Number of Measurements')
    axes[0].set_ylabel('log₁₀(Number of Branches)')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3, which='both')

    # Right: Information budget depletion
    info_used = measurements * 1  # 1 bit per measurement
    axes[1].plot(measurements, info_used, 'g-', linewidth=2, label='Information used')
    axes[1].axhline(y=max_measurements, color='red', linestyle='--', linewidth=2,
                    label=f'Max information ({max_measurements:.2e} bits)')
    axes[1].set_title('Information Budget Depletion', fontsize=12, fontweight='bold')
    axes[1].set_xlabel('Number of Measurements')
    axes[1].set_ylabel('Information (bits)')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.suptitle('Many-Worlds is FINITE: Direct Refutation of Infinite MWI', fontsize=14, fontweight='bold')
    plt.tight_layout()
    fig.savefig('finitist_many_worlds_refutation.png', dpi=300, bbox_inches='tight')
    print("  ✓ Saved: finitist_many_worlds_refutation.png")

    return fig


# Main execution
if __name__ == "__main__":
    max_measurements, log10_branches_100, log10_branches_max = finite_many_worlds()
    fig = plot_many_worlds_refutation(max_measurements, log10_branches_100, log10_branches_max)
    plt.show()
```
