```python
"""
finitist_time_evolution.py
===========================
Time as FINITE recursion (discrete steps).

Shows time is computational iteration, not continuous flow.
"""

import numpy as np
import matplotlib.pyplot as plt

# Constants
PLANCK_TIME = 5.39e-44  # s
MAX_BITS = 1e122  # Maximum information in universe

print("=" * 70)
print("TIME AS FINITE RECURSION")
print("=" * 70)

def finite_time_evolution(n_steps=1000, planck_time=PLANCK_TIME):
    """
    Simulate time as discrete computational steps.
    """
    # Discrete time steps
    time_steps = np.arange(n_steps)
    time_continuous = time_steps * planck_time

    # Simulate a quantum state evolving over time
    omega = 1e20  # Frequency (rad/s)
    state_finite = np.sin(omega * time_continuous)

    # Calculate total time
    total_time = n_steps * planck_time

    # Maximum time steps before heat death
    max_steps = int(MAX_BITS)
    max_time_years = max_steps * planck_time / (365.25 * 24 * 3600)

    print(f"\nTime Evolution:")
    print(f"  Steps simulated: {n_steps:,}")
    print(f"  Time per step: {planck_time:.2e} s (Planck time)")
    print(f"  Total time: {total_time:.2e} s")
    print(f"\nMaximum time steps (heat death): {max_steps:.2e}")
    print(f"Maximum time (years): {max_time_years:.2e}")
    print(f"\n✓ Time is FINITE recursion, not infinite continuum")
    print(f"✓ Each step is a computational update")
    print(f"✓ Time ends when entropy budget is exhausted")

    return time_steps, state_finite, total_time


def plot_time_evolution(time_steps, state):
    """Plot discrete time evolution."""
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    axes[0].plot(time_steps[:100], state[:100], 'b-', linewidth=2)
    axes[0].set_title('Quantum State Evolution\n(Discrete Time Steps)', fontsize=12, fontweight='bold')
    axes[0].set_xlabel('Time Step (n)')
    axes[0].set_ylabel('State Amplitude')
    axes[0].grid(True, alpha=0.3)

    axes[1].plot(time_steps[:10], state[:10], 'ro-', linewidth=2, markersize=8)
    axes[1].set_title('Discrete Nature of Time\n(Planck Scale Steps)', fontsize=12, fontweight='bold')
    axes[1].set_xlabel('Time Step (n)')
    axes[1].set_ylabel('State Amplitude')
    axes[1].grid(True, alpha=0.3)

    plt.suptitle('Time Is Finite Recursion, Not Continuous Flow', fontsize=14, fontweight='bold')
    plt.tight_layout()
    fig.savefig('finitist_time_evolution.png', dpi=300, bbox_inches='tight')
    print("  ✓ Saved: finitist_time_evolution.png")

    return fig


# Main execution
if __name__ == "__main__":
    time_steps, state, total_time = finite_time_evolution()
    fig = plot_time_evolution(time_steps, state)
    plt.show()
```
