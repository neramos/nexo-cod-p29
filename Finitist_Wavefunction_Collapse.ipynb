```python
"""
finitist_wavefunction_collapse.py
==================================
Finite wave function collapse.

Shows collapse is finite computation, not mystical.
"""

import numpy as np
import matplotlib.pyplot as plt

print("=" * 70)
print("FINITE WAVE FUNCTION COLLAPSE")
print("=" * 70)

def finite_wavefunction_collapse(n_states=1000):
    """
    Simulate wave function with FINITE superposition.
    """
    # Finite superposition
    n_states = min(n_states, 1000)  # Keep it manageable

    # Random amplitudes (normalized)
    amplitudes = np.random.random(n_states)
    amplitudes = amplitudes / np.linalg.norm(amplitudes)

    # Probabilities
    probabilities = amplitudes**2

    # Verify normalization
    total_prob = np.sum(probabilities)

    # Simulate collapse
    collapsed_state = np.random.choice(n_states, p=probabilities)
    collapsed_amplitude = amplitudes[collapsed_state]

    print(f"\nWave Function Properties:")
    print(f"  Number of states in superposition: {n_states:,}")
    print(f"  Total probability: {total_prob:.6f} (should be 1.0)")
    print(f"  Collapsed to state: {collapsed_state}")
    print(f"  Collapsed amplitude: {collapsed_amplitude:.6f}")
    print(f"\n✓ Wave function is FINITE superposition")
    print(f"✓ Collapse is FINITE computation (not mystical)")
    print(f"✓ Measurement extracts FINITE information")

    return n_states, amplitudes, probabilities, collapsed_state


def plot_wavefunction_collapse(amplitudes, probabilities, collapsed):
    """Plot wave function before and after collapse."""
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    states = np.arange(len(amplitudes))

    axes[0].bar(states, amplitudes, color='blue', alpha=0.5, label='Amplitude')
    axes[0].axvline(x=collapsed, color='red', linestyle='--', linewidth=2, label='Collapsed state')
    axes[0].set_title('Before Collapse: Finite Superposition', fontsize=12, fontweight='bold')
    axes[0].set_xlabel('State Index')
    axes[0].set_ylabel('Amplitude')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3, axis='y')

    collapsed_probs = np.zeros(len(probabilities))
    collapsed_probs[collapsed] = 1.0
    axes[1].bar(states, collapsed_probs, color='green', alpha=0.7)
    axes[1].set_title('After Collapse: Single Definite State', fontsize=12, fontweight='bold')
    axes[1].set_xlabel('State Index')
    axes[1].set_ylabel('Probability')
    axes[1].grid(True, alpha=0.3, axis='y')

    plt.suptitle('Wave Function Collapse: Finite Computation', fontsize=14, fontweight='bold')
    plt.tight_layout()
    fig.savefig('finitist_wavefunction_collapse.png', dpi=300, bbox_inches='tight')
    print("  ✓ Saved: finitist_wavefunction_collapse.png")

    return fig


# Main execution
if __name__ == "__main__":
    n_states, amplitudes, probabilities, collapsed = finite_wavefunction_collapse()
    fig = plot_wavefunction_collapse(amplitudes, probabilities, collapsed)
    plt.show()
```
