```python
"""
finitist_information_bound.py
==============================
Finite information in finite universe (Bekenstein bound).

Shows Hilbert space is finite, not infinite.

FIXED: Uses logarithms to avoid computational overflow
"""

import numpy as np
import matplotlib.pyplot as plt

# Constants
PLANCK_TIME = 5.39e-44  # s
MAX_BITS = 1e122  # Maximum information in observable universe

print("=" * 70)
print("FINITE INFORMATION (Bekenstein Bound)")
print("=" * 70)

def finite_hilbert_space(max_bits=MAX_BITS):
    """
    Calculate finite Hilbert space dimension.

    FIXED: Use logarithms instead of calculating 2^(10^122) directly
    """
    # DON'T calculate 2^(10^122) directly - use log10 instead
    log_max_states = max_bits * np.log10(2)  # ≈ 3e121

    print(f"\nMaximum information in observable universe:")
    print(f"  Bits: {max_bits:.2e}")
    print(f"  Distinguishable states: 2^{int(max_bits):,}")
    print(f"  ≈ 10^{log_max_states:.2e}")
    print(f"\n✓ Hilbert space is FINITE (not infinite)")
    print(f"✓ Wave function has FINITE precision")
    print(f"✓ Quantum states are COUNTABLE (not continuous)")

    return log_max_states  # Return log value, not actual number


def plot_information_bound():
    """Plot information bound comparison."""
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Left: Standard vs. Finitist
    categories = ['Standard QM', 'Finitist QM']
    axes[0].bar(categories, [100, MAX_BITS/1e120], color=['red', 'green'], alpha=0.7)
    axes[0].set_title('Hilbert Space Dimension\n(Standard vs. Finitist)', fontsize=12, fontweight='bold')
    axes[0].set_ylabel('Dimension (log scale)')
    axes[0].set_yscale('log')
    axes[0].grid(True, alpha=0.3, axis='y')

    # Right: Information growth over cosmic time
    time_cosmic = np.logspace(-44, 17, 100)
    info_growth = np.log10(time_cosmic / PLANCK_TIME)

    axes[1].loglog(time_cosmic, 10**info_growth, 'b-', linewidth=2)
    axes[1].axhline(y=MAX_BITS, color='red', linestyle='--', linewidth=2, label='Bekenstein Bound')
    axes[1].set_title('Information Growth Over Cosmic Time', fontsize=12, fontweight='bold')
    axes[1].set_xlabel('Time (s)')
    axes[1].set_ylabel('Information (bits)')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3, which='both')

    plt.suptitle('Finite Information in Finite Universe', fontsize=14, fontweight='bold')
    plt.tight_layout()
    fig.savefig('finitist_information_bound.png', dpi=300, bbox_inches='tight')
    print("  ✓ Saved: finitist_information_bound.png")

    return fig


# Main execution
if __name__ == "__main__":
    log_max_states = finite_hilbert_space()
    fig = plot_information_bound()
    plt.show()
```
