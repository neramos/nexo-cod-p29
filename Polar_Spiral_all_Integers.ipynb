```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Parameters - feel free to change
N = 20000          # Larger = more arms and detail (try 10000, 20000, 50000)
dot_size = 1.2     # Size of each dot

# Generate the spiral
theta = np.arange(1, N + 1, dtype=float)
r = theta

# Convert to Cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)

# Nice coloring for all integers (rainbow effect based on angle)
angle_color = (theta % (2 * np.pi)) / (2 * np.pi)   # normalized angle 0-1

plt.figure(figsize=(12, 12), facecolor='black')

# Plot all points with a beautiful gradient
plt.scatter(x, y,
            c=angle_color,
            cmap='hsv',      # vibrant rainbow colors
            s=dot_size,
            alpha=0.85)

plt.title("Polar Spiral of All Integers (1 to {:,})".format(N),
          color='white', fontsize=16, pad=20)

plt.axis('off')
plt.tight_layout()

# Optional: save the image
plt.savefig("integer_polar_spiral.png", dpi=400, bbox_inches='tight', facecolor='black')

plt.show()

print(f"✅ Plotted {N:,} integers in the polar spiral style")
```
