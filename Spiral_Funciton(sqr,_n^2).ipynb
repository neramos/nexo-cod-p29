```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

N = 30000   # Number of integers to plot

# Normal consecutive integers
n = np.arange(1, N+1, dtype=float)

# === theta built with square root instead of n * log(n) ===
theta = n**2
#theta = np.sqrt(n)          # This is the change you wanted
#theta = n**2*n * np.log(n + 1)
#theta = n**2 * np.log(n + 1)
r = n                       # radius remains n

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(13, 13), facecolor='black')
plt.scatter(x, y, c='#ffff00', s=1.5, alpha=0.85)

plt.title(f"Polar Spiral with θ = n^2 \n(1 to {N:,} integers)",
          color='white', fontsize=15)
plt.axis('off')
plt.tight_layout()
plt.show()
```
