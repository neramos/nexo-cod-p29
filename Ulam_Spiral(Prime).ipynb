```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def is_prime(n):
    if n < 2: return False
    if n in (2, 3): return True
    if n % 2 == 0 or n % 3 == 0: return False
    i = 5
    while i*i <= n:
        if n % i == 0 or n % (i+2) == 0: return False
        i += 6
    return True

N = 10000  # larger N shows more arms

theta = np.arange(1, N+1)
r = theta
x = r * np.cos(theta)
y = r * np.sin(theta)

colors = ['cyan' if is_prime(n) else 'yellow' for n in range(1, N+1)]

plt.figure(figsize=(10, 10), facecolor='black')
plt.scatter(x, y, c=colors, s=2, alpha=0.8)
plt.title("Polar Spiral Ulam Primes\nPrimes in cyan", color='white', fontsize=14)
plt.axis('off')
plt.tight_layout()
plt.show()
```
